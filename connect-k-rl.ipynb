```python
import os, sys, json, glob
sys.path.insert(0, ".")
```

```python
from agents.dqn import RainbowDQN
from agents.ppo import PPO
from env.connect_k import ConnectK
from eval.evaluate import run_full_eval, _agent_greedy_fn, _random_fn, _flip_board
from eval.plot import plot_learning_curves, plot_eval_results

OUT_DIR = "best_eval"
os.makedirs(OUT_DIR, exist_ok=True)

def quick_eval(agent, k, n_games=100):
    env = ConnectK(k=k)
    fn = _agent_greedy_fn(agent)
    wins = 0
    for _ in range(n_games):
        s = env.reset()
        while True:
            v = env.get_valid_actions()
            a = fn(s, v)
            nxt, r, done, info = env.step(a)
            if done:
                if info["winner"] == 1: wins += 1
                break
            opp_a = _random_fn(None, env.get_valid_actions())
            nxt2, _, done2, info2 = env.step(opp_a)
            if done2:
                break
            s = _flip_board(nxt2)
    return wins / n_games

def best_checkpoint(agent_cls, ckpts, k):
    best_path, best_rate = None, -1
    for ckpt in sorted(ckpts):
        agent = agent_cls()
        agent.load(ckpt)
        if hasattr(agent, 'online_net'):
            agent.online_net.eval()
        else:
            agent.network.eval()
        rate = quick_eval(agent, k)
        print(f"  {os.path.basename(ckpt)}: {rate:.0%}")
        if rate > best_rate:
            best_rate, best_path = rate, ckpt
    print(f"  --> Best: {os.path.basename(best_path)} ({best_rate:.0%})\n")
    return best_path

for k in [4, 5, 6]:
    print(f"\n{'='*50}")
    print(f"Finding best checkpoints for k={k}")
    print(f"{'='*50}")

    print(f"\nDQN k={k}:")
    best_dqn = best_checkpoint(RainbowDQN, glob.glob(f"checkpoints/dqn_k{k}_ep*.pt"), k)

    print(f"PPO k={k}:")
    best_ppo = best_checkpoint(PPO, glob.glob(f"checkpoints/ppo_k{k}_step*.pt"), k)

    # save which checkpoints were selected
    with open(f"{OUT_DIR}/best_ckpts_k{k}.json", "w") as f:
        json.dump({"k": k, "dqn": best_dqn, "ppo": best_ppo}, f, indent=2)

    # run full eval — monkey-patch the output path
    import eval.evaluate as _ev
    _orig = _ev.os.makedirs
    results = run_full_eval(best_dqn, best_ppo, k=k)
    # overwrite the logs/ copy with one in best_eval/
    with open(f"{OUT_DIR}/eval_k{k}_results.json", "w") as f:
        json.dump(results, f, indent=2)

    # plots
    dqn_log = f"logs/dqn_k{k}_log.json"
    ppo_log = f"logs/ppo_k{k}_log.json"
    if os.path.exists(dqn_log) and os.path.exists(ppo_log):
        plot_learning_curves(dqn_log, ppo_log, f"{OUT_DIR}/learning_curves_k{k}.png")
    plot_eval_results(f"{OUT_DIR}/eval_k{k}_results.json", f"{OUT_DIR}/eval_bar_k{k}.png")

print(f"\nAll outputs saved to {OUT_DIR}/")
```
